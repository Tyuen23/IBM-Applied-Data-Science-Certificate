```python
# Gets the name of the category

def get_category_type(row):
    try:
        categories_list = row['Category']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
radius = 500
LIMIT = 100

venues = []

for lat, long, post, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['PostalCode'], df['Borough'], df['Neighbourhood']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    for venue in results:
        venues.append((
            post, 
            borough,
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
v_df = pd.DataFrame(venues)
v_df.columns = ['PostalCode', 'Borough', 'Neighbourhood', 'BoroughLatitude', 'BoroughLongitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']
print(v_df.shape)
v_df.head()
```

```python
# one hot encoding
yyz_central_onehot = pd.get_dummies(v_df[['VenueCategory']], prefix="", prefix_sep="")

# add postal, borough and neighborhood column back to dataframe
yyz_central_onehot['PostalCode'] = v_df['PostalCode'] 
yyz_central_onehot['Borough'] = v_df['Borough'] 
yyz_central_onehot['Neighbourhoods'] = v_df['Neighbourhood'] 

# move postal, borough and neighborhood column to the first column
fixed_columns = list(yyz_central_onehot.columns[-3:]) + list(yyz_central_onehot.columns[:-3])
yyz_central_onehot = yyz_central_onehot[fixed_columns]

print(yyz_central_onehot.shape)
yyz_central_onehot.head()
```

```python
yyz_central_freq = yyz_central_onehot.groupby(['PostalCode', 'Borough', 'Neighbourhoods']).mean().reset_index()
print(yyz_central_freq.shape)
yyz_central_freq.head()
```

```python
top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
areaColumns = ['PostalCode', 'Borough', 'Neighbourhoods']
freqColumns = []
for ind in np.arange(top_venues):
    try:
        freqColumns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        freqColumns.append('{}th Most Common Venue'.format(ind+1))
columns = areaColumns+freqColumns

# create a new dataframe
venues_sorted = pd.DataFrame(columns=columns)
venues_sorted['PostalCode'] = yyz_central_freq['PostalCode']
venues_sorted['Borough'] = yyz_central_freq['Borough']
venues_sorted['Neighbourhoods'] = yyz_central_freq['Neighbourhoods']

for ind in np.arange(yyz_central_freq.shape[0]):
    row_categories = yyz_central_freq.iloc[ind, :].iloc[3:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    venues_sorted.iloc[ind, 3:] = row_categories_sorted.index.values[0:top_venues]

venues_sorted.sort_values(freqColumns, inplace=True)
venues_sorted
```

```python
kclusters = 5

yyz_central_freq_cluster = yyz_central_freq.drop(['PostalCode', 'Borough', 'Neighbourhoods'], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(yyz_central_freq_cluster)

yyz_central_clustered_df = df
yyz_central_clustered_df['Cluster'] = kmeans.labels_

yyz_central_clustered_df = yyz_central_clustered_df.join(venues_sorted.drop(['Borough', 'Neighbourhoods'], 1).set_index('PostalCode'), on='PostalCode')
yyz_central_clustered_df.sort_values(['Cluster'] + freqColumns, inplace=True)
yyz_central_clustered_df
```

```python
# create map
yyz_clusters = folium.Map(location=[yyz_lat, yyz_lng], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, post, bor, poi, cluster in zip(yyz_central_clustered_df['Latitude'], yyz_central_clustered_df['Longitude'], yyz_central_clustered_df['PostalCode'], yyz_central_clustered_df['Borough']
                                             ,yyz_central_clustered_df['Neighbourhood'], yyz_central_clustered_df['Cluster']):
    label = folium.Popup('{} ({}): {} - Cluster {}'.format(bor, post, poi, cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(yyz_clusters)
       
yyz_clusters
```
